```python
import os
from pathlib import Path
import pandas as pd

# defining paths for notebook
current_dir = Path(globals()["_dh"][0])

raw_data_path = os.path.join(current_dir, r"..\..\raw(bronze)")
processed_data_path = os.path.join(current_dir, "modalities.csv")
```

```python
df_all_path = os.path.join(raw_data_path, "events_history.csv")
df_all = pd.read_csv(df_all_path, usecols=["sport"])
```

```python
df_paris_path = os.path.join(raw_data_path, "events_paris.csv")
df_paris = pd.read_csv(df_paris_path, usecols=["sport"])
```

## Testing DataFrames

# Processing dataframes

```python
df_final = pd.concat([df_all, df_paris])
df_final = df_final.drop_duplicates()
df_final = df_final.sort_values(by="sport")
df_final.head(10)
```

# Saving to csv

```python
df_final.to_csv(processed_data_path, index=False)
```
