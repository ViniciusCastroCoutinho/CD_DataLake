```python
import os
from pathlib import Path
import pandas as pd

# defining paths for notebook
current_dir = Path(globals()["_dh"][0])

raw_data_path = os.path.join(current_dir, r"..\..\raw(bronze)")
processed_data_path = os.path.join(current_dir, "athletes_per_gender.csv")
parquet = os.path.join(current_dir, "athletes_per_gender.parquet")
```

```python
df_all_path = os.path.join(raw_data_path, "athletes_history.csv")
df_all = pd.read_csv(df_all_path)
```

```python
df_paris_path = os.path.join(raw_data_path, "athletes_paris.csv")
df_paris = pd.read_csv(df_paris_path)
```

# Processing Dataframes

## Processing athletes history

```python
df1 = df_all["sex"].value_counts()
df1 = pd.DataFrame([df1]).T
df1.index.names = ["gender"]
df1.head()
```

```python
df_all = df1
```

## Processing athletes in Paris 2024

```python
df2 = df_paris["gender"].value_counts()
df2 = pd.DataFrame([df2]).T
df2.head()
```

```python
df_paris = df2
```

## Merging DataFrames

```python
df_final = pd.concat([df_all, df_paris])
df_final = df_final.groupby(["gender"]).sum()
df_final.head()
```

## Saving files

```python
df_final.to_csv(processed_data_path)
df_final.to_parquet(parquet, index=False, engine="fastparquet")
```
