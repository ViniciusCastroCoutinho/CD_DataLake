```python
import os
from pathlib import Path
import pandas as pd

# defining paths for notebook
current_dir = Path(globals()["_dh"][0])

raw_data_path = os.path.join(current_dir, "..")
```

```python
files = [
    "athletes_history",
    "athletes_paris",
    "events_history",
    "events_paris",
    "olympics_history",
    "olympics_paris2024",
]
```

```python
for file in files:
    df_path = os.path.join(raw_data_path, file + ".csv")
    df = pd.read_csv(df_path, low_memory=False)

    parquet = os.path.join(current_dir, file + ".parquet")
    df.to_parquet(parquet, index=False, engine="fastparquet")
```
